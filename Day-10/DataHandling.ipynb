## Create the List of Patients, Those are in ICU

```python
import pandas as pd

data = pd.read_csv("patients.csv")
data.columns = data.columns.str.strip()

print("Column Names:", data.columns.tolist())
print("\nUnique service names:", data['service'].unique())


icu_patients = data[data['service'].str.lower() == 'icu']
print(f"\nICU Patient Records: {icu_patients.shape[0]} rows and {icu_patients.shape[1]} columns")

pd.set_option('display.max_columns', None)
print("\nICU Patients Data:\n", icu_patients)

icu_patients.to_csv("icu_patients.csv", index=False)
print("\nFile 'icu_patients.csv' created successfully!")
```
