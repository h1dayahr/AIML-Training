```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 40]

plt.plot(x, y)
plt.show()
```

### Basics of MatPlotLib Chart
   1. PyPlot: Interface for Creating and Managing Plot
   2. Figure: The Complete Window to Create Plot
   3. Axes: The Area Where Data is Ploted
   4. Axis: x Line and y Line to Define Scale and Ticks
   5. Plot: The Graph of Data

```python
plt.figure(figsize=(8,6))
```

```python
plt.plot(x,y,marker='o',linestyle='-')
```

```python
plt.plot(x,y,marker='.',linestyle='-')
```

```python
plt.plot(x,y,marker=',',linestyle='-')
```

### First Example to Create a Line Chart From Our Data
      1. Students.xlsx for fetching the data
      2. Generate Line Chart Based on Student Name and Student Marks

```python
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/Nurhidayah R/OneDrive/Desktop/AIML-Training/Day-10/students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='s',color='yellow')
plt.title('Students Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(names,marks,marker='s',color='pink')
plt.title('Students Progress Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='s',color='pink')
plt.title('Students Progress Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='s',color='pink')
plt.title('Students Progress Chart')
plt.xlabel('---Name---')
plt.ylabel('---Marks---')
plt.grid(True)
plt.show()
```

```python
print('Solid Line')
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='s',color='pink')
plt.title('Students Progress Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o',color='blue',linestyle='--')
plt.show()
```

```python
print('Dotted Line')
plt.plot(names,marks,marker='o',color='green',linestyle=':')
plt.show()
```

```python
print('Dashed Dotted Line')
plt.plot(names,marks,marker='o',color='red',linestyle='-.')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='s',color='green',linestyle='-.')
plt.title('Students Progress Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Solid Line')
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='h',color='blue')
plt.title('Students Progress Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Solid Line')
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='v',color='green')
plt.title('Students Progress Chart')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```
