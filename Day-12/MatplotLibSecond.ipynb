### Create a Bar Plot-Average Age of Patients by Arrival Date

```python
df['arrival_date'] = pd.to_datetime(df['arrival_date'])
print(df['arrival_date'])
avg_age_by_date=df.groupby('arrival_date')['age'].mean()
print(avg_age_by_date)
```

```python
plt.figure(figsize=(10,4))
plt.bar(avg_age_by_date.index, avg_age_by_date.values, color='lightcoral', edgecolor='black')
plt.title('Average Age of Patients by Arrival Date')
plt.xlabel('Arrival Date')
plt.ylabel('Average Age')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
players=['Dd','Irfan','Joe','Max','James']
match1_score=[45,67,56,78,89]
match2_score=[55,78,40,80,97]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match One', s=100)
plt.scatter(players,match2_score,color='green',label='Match Two', s=100, marker="x")
plt.title('Comparision of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.show()
```

```python
players=['Dd','Irfan','Joe','Max','James']
match1_score=[45,67,56,78,89]
match2_score=[55,78,40,80,97]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match One', s=100, marker="d")
plt.scatter(players,match2_score,color='green',label='Match Two', s=100, marker="x")
plt.title('Comparision of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

### Write a comparision that shows average calories burned for each workout type from final_data dataset

```python
df=pd.read_csv('Final_data.csv')
avg_cal=df.groupby('Workout_Type')['Calories_Burned'].mean()
selected_index=avg_cal.index
selected_value=avg_cal.values
plt.scatter(selected_index,selected_value,color='green', s=100,marker='p')
plt.title('Comparision of Workout and calories Burned')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.grid(True,linestyle='-',alpha=0.5)
plt.show()
```

### Create a Bar Chart using Following Data
1. python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
### Liked by Users as Follows
2. liked_by=[5000,7000,9000,4000]
### Create a Vertical Bar Graph where Each Bar Represents The Value of python_lib Respective liked by users
3. Every Bar must have Different Color

```python
python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
liked_by=[5000,7000,9000,4000]
bar_colors=['pink','skyblue','yellow','grey','purple']
plt.figure(figsize=(4,2))
plt.bar(python_lib,liked_by,color=bar_colors,width=0.25,edgecolor='black')
plt.title('Python Library Likes')
plt.xlabel('Library')
plt.ylabel('Likes')
plt.show()
```
