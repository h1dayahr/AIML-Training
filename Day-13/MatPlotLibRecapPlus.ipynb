### Bar Plot Example
1. plt.bar(x,y,color='',edgecolor=' ')

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['pink','skyblue','beige','purple','teal']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Categories wise Sales Report')
plt.grid(axis='y',linestyle='--',color='green',alpha=0.5)
plt.tight_layout()
plt.show()

# Alpha is for Visibility of Grid Lines
```

### Scatter Plot
1.plt.scatter(x,y,s='',c='colorName',marker='makername',alpha=transp.,edgecolor='color name',label='lable of legend')

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg,s=100,c='pink',marker='o',alpha=1,edgecolor='black',label='weight/height')
plt.title('Scatter Plote Height Vs Weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.xlim(140,190)
plt.ylim(40,90)
plt.xticks(rotation=45, fontsize=12, color='black')
plt.yticks([50,55,60,65,70,75,80,85],['50Kg','55kg','60Kg','65Kg','70kg','75kg','80kg','85kg'],fontsize=12, color='black')
plt.grid(color='lightblue',alpha=0.7,linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
# Function                | What It Does                     | Example                        |
# ----------------------- | -------------------------------- | ------------------------------ |
# `xlabel()` / `ylabel()` | Adds axis labels                 | Height (cm), Weight (kg)   |
# `xlim()` / `ylim()`     | Sets the visible range of data   | `plt.xlim(140,190)`            |
# `xticks()` / `yticks()` | Customizes tick marks and labels | Label ticks as 50kg, 55kg |

# plt.scatter(xaxis, yaxis)
# The first argument  height_cm  goes on the x-axis
# The second argument  weight_kg  goes on the y-axis

# alpha= is for Visibility of Grid Lines

# plt.legend() is use to show a small box on your graph that explains what each color, shape, or line represents.
```
