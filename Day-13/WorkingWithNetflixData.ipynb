```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=30,color='pink', edgecolor='teal', alpha=0.8) 
plt.title('Netflix realeased Titles by Year ')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='black',linestyle='--',linewidth=0.7)   
plt.tight_layout()
plt.savefig('netflix_bar.pdf',dpi=300,bbox_inches='tight')
plt.show()  
```

```python
movies=df[df['type']=='Movie']
print('Number of Movies: \t ',movies['show_id'].count())
shows=df[df['type']=='TV Show']
print('Number of TV Shows: \t ',shows['show_id'].count())
print()
total=movies['show_id'].count() + shows['show_id'].count()
print('Total of Movies and TV Shows: \t ', total)
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'], shows['release_year']], 
         bins=30,
         color=['pink','skyblue'],
         edgecolor='magenta',
         alpha=0.8,
         label=['Movies','TV-Shows']
        ) 
plt.title('Netflix realeae by Year and Type ')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid( axis='y', color='lightblue',linestyle='--',linewidth=0.7)   
plt.tight_layout()
plt.legend(title='Movies/shows')
plt.savefig('netflix_sbowvsmovies.jpg',dpi=300,bbox_inches='tight')
plt.show()
```
