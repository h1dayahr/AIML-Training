### Matplotlib.pyplot.subplots()
1. plt.subplots() is a function in Matplotlib that helps you create multiple plots (subplots) inside one single figure (a figure is like a "canvas" or "page" for your plots)
2. So instead of drawing only one chart, you can show many small charts together — for example, a 2x2 grid of plots.

```python
# fig, ax = plt.subplots(nrows, ncols)
# Parameters:
# nrows: number of rows of plots
# ncols: number of columns of plots

# It returns a tuple:
# fig → the entire figure (the overall canvas)
# ax → the axes (the individual plots)
```

```python
import matplotlib.pyplot as plt
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6),(ax7,ax8))=plt.subplots(nrows=4,ncols=4,figsize=(16,10))
```

```python
import numpy as np
```

```python
sales_data={'Electronics':45000,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
actual_sale=list(sales_data.values())
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/4)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(category,actual_sale)
ax4.hist(np.random.randn(1000))
```

### Subplots Using Indexing

```python
our_lines=np.linspace(0,10,100)
```

```python
product_prices={'Hair Oil':20, 'Face Wash':30,'Lip Balm':10,'Face Cream':15}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(1000)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['pink','purple','beige','teal'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['pink','purple','beige','teal'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='pink')
plt.tight_layout()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['pink','purple','beige','teal'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='pink')
ax[1,1].hist(our_dots,color='teal')
# first row 0 & first column 0 means [0,0]  Line Plot
#-----------------------------------------------------------------
ax[0,0].set_title('Line Plot Example')
ax[0,0].set_xlabel('X Axis')
ax[0,0].set_ylabel('Y Axis')
ax[0,0].grid(True,linestyle='--',alpha=0.7, color='lightblue',axis='y')
# first row 0 & second column 1 means [0,1]  Line Plot
#---------------------------------------------------------------------
ax[0,1].set_title('Product Sales Information ')
ax[0,1].set_xlabel('Products Name')
ax[0,1].set_ylabel('Product Sale')
ax[0,1].grid(True,linestyle='--',alpha=0.7, color='red',axis='y')
#---------------------------------------------
#-------------Second row 1 and first column 0 means [1,0]
ax[1,0].set_title('Our Scatter Plot ')
ax[1,0].set_xlabel('Products Name')
ax[1,0].set_ylabel('Product Sale')
ax[1,0].grid(True,linestyle='--',alpha=0.7, color='lightblue')
#-------------Second row 1 and second column 1 means [1,1]
ax[1,1].set_title('Histogram Example')
ax[1,1].set_xlabel('Value Range')
ax[1,1].set_ylabel('Frequency')
ax[1,1].grid(True,linestyle='--',alpha=0.7, color='pink')
#-----------------------------------------------------------------
fig.suptitle('Matplot Lib SubPlots example using 2 rows and 2 cols')
#-----------------------------------------------------------------------
plt.tight_layout()
plt.savefig('subplotex.jpeg',dpi=300,bbox_inches='tight')
```
