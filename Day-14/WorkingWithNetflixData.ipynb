```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/Nurhidayah R/OneDrive/Desktop/AIML-Training/Day-13/netflix_data.csv')
df.head(3)
```

```python
sns.set_theme(style='whitegrid',palette='pastel')
sns.countplot(data=df,x='type')
plt.title('Number of Movies Vs Number of TV Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df.head(10),x='type',hue='country', palette='Set2')
plt.title('Netflix Conternt Type by Country only top 10')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country')
plt.show()     
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df.head(50),x='type',hue='country', palette='pastel')
plt.title('Netflix Conternt Type by Country only top 10')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country',bbox_to_anchor=(1.02,1),loc='upper left',frameon=True)
plt.tight_layout()
plt.show() 
```

### Use Netflix dataset to visualize the distribution of release years for Movies Vs TV Shows using Seaborn's histplot()
1. Use different color for Movie Vs TV Show

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='release_year', hue='type', palette='pastel')
plt.title('Distribution of Release Years: Movies vs TV Shows')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
