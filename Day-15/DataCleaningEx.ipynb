#### Quick Summary

🧩 1. isnull()
→ Checks if data is missing (NaN or None).
Example:
    df.isnull()
Shows True where data is missing, False where data exists.

To count how many missing values each column has:
    df.isnull().sum()

---

🧹 2. dropna()
→ Removes rows (or columns) that have missing data.
Example:
    df.dropna()
Deletes any row that contains at least one missing value.

To remove rows only when certain columns are missing:
    df.dropna(subset=['gender'])

---

🧴 3. fillna()
→ Fills (replaces) missing data with a value.
Example:
    df['study_hours'].fillna(5)
Fills missing values in “study_hours” with 5.

To fill with the average (mean) of the column:
    df['study_hours'].fillna(df['study_hours'].mean())

To fill multiple columns at once:
    df.fillna({'study_hours': df['study_hours'].mean(), 'age': 20})

---

📊 4. notnull()
→ Opposite of isnull(), checks where data exists.
Example:
    df.notnull()
Shows True where data exists, False where missing.

---

💡 Summary Table:

| Function   | What It Does          | Example                     | Meaning                      |
|-------------|----------------------|------------------------------|-------------------------------|
| isnull()    | Finds missing values | df.isnull()                  | True for missing              |
| notnull()   | Finds non-missing    | df.notnull()                 | True where data exists        |
| dropna()    | Removes missing rows | df.dropna()                  | Deletes missing rows          |
| fillna()    | Replaces missing     | df.fillna(0)                 | Fills missing values with 0   |
