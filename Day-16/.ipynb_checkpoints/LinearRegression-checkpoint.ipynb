##### 1. Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data=pd.read_csv('sales_data.csv')
data.head()
```

```python
X=data[['Advertising_Spend']]
y=data['Sales_Revenue']
```

```python
model=LinearRegression()
model.fit(X,y)
print('Training Done')
```

```python
print()
new_spend=float(input('Enter New Spend to Predict Sales Revenue: \t '))
predicted_revenue=model.predict([[new_spend]])
print(f'Predicted Sales Revenue is RM {predicted_revenue} for New Spend : \t {new_spend}')
```
