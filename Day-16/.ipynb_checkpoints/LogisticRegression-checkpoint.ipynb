###### 2. Logistic Regression - Pass/Fail

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('pass_fail.csv')
df.head(7)
```

```python
# Split The Data
X=df[['Hours']]
y=df['Result']
```

```python
# Create The Model
model=LogisticRegression()
```

```python
# Train The Model
model.fit(X,y)
```

```python
pred_35=model.predict([[3.5]])
print('Predicted Result for 3.5 Hours:',pred_35[0])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('customer_purchase.csv')
df.head()
```

```python
le=LabelEncoder()
df['Purchased']=le.fit_transform(df['Purchased'])
X=df[['Age','Salary']]
y=df['Purchased']
```

```python
df
```

```python
model=LogisticRegression()
```

```python
model.fit(X,y)
```

```python
new_customer=[[28,35000]]
prediction=model.predict(new_customer)
print(f"Predicted Result (1=Yes, 0=No'): {prediction[0]}")
```

```python
# Prediction 
new_cust_age=int(input('Enter Age:'))
new_cust_sal=int(input('Enter Salary:'))
new_cust=[[new_cust_age,new_cust_sal]]
prediction=model.predict(new_cust)
if (prediction):
    print('Purchased')
else:
    print('Not Purchased')
```

###### Exercise - Logistic Regression Spam/NotSpam
1. Download spam_data.csv
2. Load the Data
3. Do The Data Cleaning and Labelling
4. Use Logistic Regression to Trained the Model
5. Predict Result if The Email is Spam or Not Spam

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('spam_dataset.csv')
df.head()
```

```python
X=df['message_content']
y=df['is_spam']
```

```python
cv=CountVectorizer()
X_cv=cv.fit_transform(X)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_cv,y)
```

```python
msg = input("Enter a message to check: ")

if msg in df['message_content'].values:
    msg_cv = cv.transform([msg])
    pred = model.predict(msg_cv)
    if pred[0] == 1:
        print("This message is SPAM (found in dataset)")
    else:
        print("This message is NOT SPAM (found in dataset)")
else:
    print("Message not found in dataset!")
```
