##### KNN Algorithm in Supervised Machine Learning
##### KNeighborsClassifier of sklearn Library 

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=np.array([
    [1,3500],
    [2,3200],
    [3,3000],
    [4,2500],
    [5,2300],
    [6,2000],
    [7,2200],
    [2,3700],
])
y=['Unfit','Unfit','Unfit','Fit','Fit','Fit','Fit','Unfit']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
# Create The Model
model.fit(X,y)
```

```python
# Predict
new_person=[[4.5,7600]]
pred=model.predict(new_person)
print('Predicted Result:',pred[0])
```

```python
# Prediction from Taking the Values from Users
exercise_hours=float(input('Enter Number of Exercise Hours: \t '))
cal_intake=float(input('Enter Number of Calories Consumed: \t '))
pred_fu=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories Intake {cal_intake} Your are like {pred_fu[0]}')
```
