##### **1. MAE: Mean Absolute Error**

```python
from sklearn.metrics import mean_absolute_error
```

```python
actual_values=[70,80,90,85,95]
predicted_values=[68,82,88,80,97]
```

```python
mae=mean_absolute_error(actual_values,predicted_values)
print('MAE',mae)
```

##### **2. MSE: Mean Squarred Error**

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
house_size=np.array([[1000],[1500],[2000],[2500],[3000]])
house_prize_thousand=np.array([120,180,250,310,370])
```

```python
model=LinearRegression()
```

```python
model.fit(house_size,house_prize_thousand)
```

```python
y_pred=model.predict(house_size)
```

```python
mse=mean_squared_error(house_size,y_pred)
```

```python
print('MSE for This Model: \t ',mse)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(house_size,house_prize_thousand,color='pink',label='Actual Values',s=100)
plt.plot(house_size,y_pred,color='teal',label='Predicted Values') 
plt.grid(color='black',linestyle='--',linewidth=0.8)
plt.xlim(750, 3250) 
plt.ylim(50, 400) 
plt.show()
```

##### **3. RSME: Root Mean Squared Error**

```python
actual_values=[22,25,30,35,40]
predicted_values=[22,26,28,29,36]
rmse=mean_squared_error(actual_values,predicted_values)
print('Mean Squared Error: \t ',rmse)
```

```python
mse=mean_squared_error(house_size,y_pred)
print(mse)
```

```python
rmse=mean_squared_error(actual_values,predicted_values)
print('Root Mean Squared Error: \t ',rmse)
```

```python
rmse=np.sqrt(mse)
print('Root Mean Squared Error: \t ',rmse)
```
