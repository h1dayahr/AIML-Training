**Matrix Multiplication We will use tf.matmulO method Rules**
1. Inner dimensions must match (m,n)@(n.p) for e.g.
*   (5,7)@(8,9) It will not work because n are different
*   (5,8)@(8,9) It will work because inner dimension is same
2. The result matrix has the shape of outer dimension, means A (m,n) B (n,p) So result has (m,p) e.g (4,7)@(7,5)-> (4,5)






```python
import tensorflow as tf
import numpy as np
```

```python
tensor_A=tf.constant([[1,2],[3,4],[4,5]],dtype=tf.int32)
tensor_B=tf.constant([[5,6],[7,8]],dtype=tf.int32)
print('Shape of A',tensor_A.shape)
print('Shape of B',tensor_B.shape)
```

```python
tensor_C=tf.matmul(tensor_A,tensor_B)
print(tensor_C)
print('Shape of tensor_C:',tensor_C.shape)
```

```python
X=tf.constant([[1.,0.,-1.],
                [2.,3.,1.]],
                 dtype=tf.float32)
X
```

Transpose: Transpose swaps axes. For 2D matrix shape(m.n) become(n,m) after transpose

```python
tensor_one=tf.constant([
    [[1,2],[2,3]],
    [[4,5],[6,7]]
],dtype=tf.int32)
tensor_one
```

```python
tensor_one_transpose=tf.transpose(tensor_one)
print(tensor_one_transpose)
print('Shape of tensor_one_transpose',tensor_one_transpose.shape)
```

```python
# 3D Transpose with Explicit axis Perrmutation
tensor_two=tf.constant([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
],dtype=tf.int32)
print(tensor_two)
```

```python
# Swap First and Last axis (0,1,2)->(0,2,1)
t2_transpose=tf.transpose(tensor_two,perm=[0,2,1])
print(t2_transpose)
print('\nShape of t2_transpose',t2_transpose.shape)
print('\nShape of tensor_twi',tensor_two.shape)
```

```python
# Reshape
#tf.reshape(tensor, new_shape)
# Change the shape without Changing the Data
# Rule: The Total Number of Elements must Stay The Same
```

```python
tensor=tf.constant([
    [1,2,3],
    [4,5,6]
],dtype=tf.int32)
print(tensor)
print('\nShape of Tensor',tensor.shape)
print('\nNumber of Elements are:',tf.size(tensor).numpy())
```

```python
tensor_r1=tf.reshape(tensor,shape=(3,2))
print(tensor_r1)
print('\nShape of tensor_r1',tensor_r1.shape)
print('\nNumber of elements are:',tf.size(tensor_r1).numpy())
```

```python
tensor_r1=tf.reshape(tensor,shape=(1,6))
print(tensor_r1)
print('\nShape of tensor_r1',tensor_r1.shape)
print('\nNumber of elements are:',tf.size(tensor_r1).numpy())
```

```python
# Infer Dimension
tensor_r4=tf.reshape(tensor,shape=(-1,2))
print(tensor_r4)
print('\nShape of tensor_r4',tensor_r4.shape)
print('\nNumber of elements are:',tf.size(tensor_r4).numpy())
```
