```python
# Tensor: An n-Dimensional Array of Numbers (n=0,n=1,n=2......)
tensor=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
tensor
```

```python
# Find the **Shape**, **Rank**, **Dimension** and **Size** of The

#SCALAR (Single Number)
print("----Scalar Information----")
print("Shape: \t",scalar.set_shape)
print("\nDimension: \t",scalar.ndim)
print("\nRank: \t",tf.rank(scalar).numpy())
print("\nSize: \t",tf.size(scalar))
```

```python
#VECTOR (1D Array)
print("----Vector Information----")
print("Shape: \t",vector.shape)
print("\nDimension: \t",vector.ndim)
print("\nRank: \t",tf.rank(vector).numpy())
print("\nSize: \t",tf.size(vector))
```

```python
#MATRIX (2D Array)
print("----Matrix Information----")
print("Shape: \t",matrix.shape)
print("\nDimension: \t",matrix.ndim)
print("\nRank: \t",tf.rank(matrix).numpy())
print("\nSize: \t",tf.size(matrix))
```

```python
#TENSOR (3D Array)
print("----Tensor Information----")
print("Shape: \t",tensor.shape)
print("\nDimension: \t",tensor.ndim)
print("\nRank: \t",tf.rank(tensor).numpy())
print("\nSize: \t",tf.size(tensor))
```

```python
import numpy as np
numpy_Array=np.arange(1,25,dtype=np.int16)
numpy_Array
```

```python
tensor_numpy=tf.constant(numpy_Array,shape=(2,3,4))
tensor_numpy
```

```python
print("----Tensor Information----")
print("Shape: \t",tensor_numpy.shape)
print("\nDimension: \t",tensor_numpy.ndim)
print("\nRank: \t",tf.rank(tensor_numpy).numpy())
print("\nSize: \t",tf.size(tensor_numpy))

print("\n\nDataType of Elements: \t",tensor_numpy.dtype)
print("\nElement along axis 0: \t",tensor_numpy.shape[0])
print("\nElement along axis 1: \t",tensor_numpy.shape[1])
print("\nElement along axis 2: \t",tensor_numpy.shape[2])
print("\nTotal Number of Elements: \t",tf.size(tensor_numpy).numpy())
```

**Arithmetic Operation with Tensors**


*   Add
*   Sub
*   Multi
*   Div

```python
tensor_one=tf.constant([[11,9],[2,5]])
tensor_one
```

```python
# tensor_one+51
tf.add(tensor_one,51)
```

```python
tensor_one-5
tf.substract(tensor_one,5)
```

```python
tensor_one*2
tf.multiply(tensor_one,2)
```

```python
tf.divide(tensor_one,2)
```

**Exercise**


*   Create two tensors with random values between 0 and 1, shape is [6,300].
*   Find out the following details: Number of elements (Size, Rank Shape)



```python
import tensorflow as tf
import numpy as np
```

```python
tensor1 = tf.random.uniform(shape=[6, 300], minval=0, maxval=1)
tensor2 = tf.random.uniform(shape=[6, 300], minval=0, maxval=1)
```

```python
print("Tensor 1:",tensor1)
print("Shape:", tensor1.shape)
print("Rank (number of dimensions):", tf.rank(tensor1).numpy())
print("Size (number of elements):", tf.size(tensor1).numpy())

print("\nTensor 2:",tensor2)
print("Shape:", tensor2.shape)
print("Rank (number of dimensions):", tf.rank(tensor2).numpy())
print("Size (number of elements):", tf.size(tensor2).numpy())
```
